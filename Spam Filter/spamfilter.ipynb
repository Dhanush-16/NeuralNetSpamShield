```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
token=Tokenizer(num_words=5000)
```

```python
import pandas as pd
```

```python
df=pd.read_csv("spam.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encod=LabelEncoder()
```

```python
y=encod.fit_transform(df["Category"])
```

```python
dic=5000
```

```python
token.fit_on_texts(df["Message"])
```

```python
sequences = token.texts_to_sequences(df["Message"])
```

```python
print(sequences)
```

```python
length=max([len(i) for i in sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
l=pad_sequences(sequences,maxlen=length,padding="pre")
```

```python
from keras.models import Sequential
from keras.layers import Dense,Embedding,LSTM
```

```python
model=Sequential()
```

```python
model.add(Embedding(dic,100,input_length=length))
model.add(LSTM(100))
model.add(Dense(units=1,activation="sigmoid",kernel_initializer="glorot_uniform"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(l,y,epochs=15,batch_size=10,validation_split=0.2)
```

```python
model.save("spam_model.h5")
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
import joblib
joblib.dump(token,"token.pkl")
```

```python
from tensorflow.keras.models import load_model
model = load_model("spam_model.h5")
```

```python
token=joblib.load("token.pkl")
```

```python
def predict(text):
    t=token.texts_to_sequences([text])
    t=pad_sequences(t,maxlen=183)
    return model.predict(t)
```
